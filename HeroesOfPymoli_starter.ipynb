```python
# Dependencies and Setup
import pandas as pd
from numpy import mean

# File to Load (Remember to Change These)

file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Pull all records of SN and count the length
players_df = purchase_data.loc[:,"SN"]
total_players = len(players_df.unique())

# Create data frame with the first index being the total players 
player_count__df = pd.DataFrame(
    {"Total Players":[total_players]
    })
player_count__df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# NUMBER OF UNIQUE ITEMS
number_of_unique_items = len(purchase_data.loc[:,"Item Name"].unique().tolist())

# AVERAGE PRICE
average_price_float = mean(purchase_data.loc[:,"Price"].tolist())
average_price = '${:.2f}'.format(average_price_float)

#NUMBER OF PURCHASES
number_of_purchases = len(purchase_data.loc[:,"Purchase ID"].tolist())

#TOTAL REVENUE
total_revenue_num = average_price_float * number_of_purchases
total_revenue = '${:,.2f}'.format(total_revenue_num)

# CREATE TABLE
purchasing_analysis_total_df = pd.DataFrame(
    {'Number of Unique Items':[number_of_unique_items],
     'Average Price':[average_price],
     'Number of Purchases':[number_of_purchases],
     'Total Revenue':[total_revenue]
    })
purchasing_analysis_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create list of unique genders in Gender column
gender_list = purchase_data.Gender.unique()

# Create list where gender, count, and percentage are stored for each player
count_list = []
percentage_list = []


for gender in gender_list:
    
    # Create data frame for gender unique values (e.g. Male only records, Female only records, etc)
    gender_df = purchase_data.loc[purchase_data["Gender"]==gender,:]
    
    # Create series of unique screen names for each gender
    unique_gender_sn = gender_df.loc[:,"SN"].unique()
    
    # TOTAL COUNT
    # Count total unique screen names 
    gender_count = len(unique_gender_sn)
    count_list.append(gender_count)
    
    # PERCENTAGE OF PLAYERS
    # Calculate percentage of each player and format to %
    percentage_of_players = f'{round(100*gender_count/total_players,2)}%'
    percentage_list.append(percentage_of_players)
    #total_gender_count.append([gender,gender_count,f'{percentage_of_players}%'])
    
# Create final data frame and set index to genders
gender_demo_df = pd.DataFrame({
    "Total Count":count_list, "Percentage of Players":percentage_list
}, index = gender_list)


# Sort data frame 

gender_demo_df.sort_values(by="Total Count",ascending=False)
gender_demo_df 


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# GENDERS
genders_list = purchase_data.groupby("Gender").count().index.values


# PURCHASE COUNT
purchase_count_list = purchase_data.groupby("Gender").count().loc[:,"Purchase ID"].tolist()

# AVERAGE PURCHASE PRICE
average_purchase_price_list = purchase_data.groupby("Gender").mean().loc[:,"Price"].tolist()

# TOTAL PURCHASE VALUE
x = 0
total_purchase_value_list = []
while x < len(average_purchase_price_list):
    total_purchase_value_list.append(average_purchase_price_list[x] *  purchase_count_list[x])
    x +=1
    
    
# CREATE DATA FRAME
purchase_analysis_df = pd.DataFrame({
    "Gender":genders_list,
    "Purchase Count":purchase_count_list,
    "Average Purchase Price":average_purchase_price_list,
    "Total Purchase Value":total_purchase_value_list,
})

purchase_analysis_df = purchase_analysis_df.set_index("Gender")
    

# AVG TOTAL PURCHASE PER PERSON    
avg_total_purchase_per_person_list = []

for gender in genders_list:
    gender_total_count = gender_demo_df.loc[gender,"Total Count"]
    gender_total_purchase_value = purchase_analysis_df.loc[gender,"Total Purchase Value"]
    gender_avg_total_purchase_per_person = gender_total_purchase_value / gender_total_count  
    avg_total_purchase_per_person_list.append(gender_avg_total_purchase_per_person)
    

# Add avg total purchase per person into dataframe
purchase_analysis_df["Avg Total Purchase per Person"] = avg_total_purchase_per_person_list

# FORMATTING
purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].apply('${:.2f}'.format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].apply('${:.2f}'.format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].apply('${:.2f}'.format)

purchase_analysis_df
```

## Age Demographics

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# UNIQUE SN
unique_sn_list = purchase_data.groupby("SN").count().index.values


# PURCHASE COUNT
purchase_count_list = purchase_data.groupby("SN").count().loc[:,"Purchase ID"].tolist()


# AVERAGE PURCHASE PRICE
average_purchase_price_list = purchase_data.groupby("SN").mean().loc[:,"Price"].tolist()


# TOTAL PURCHASE VALUE
x = 0
total_purchase_value_list = []
while x < len(average_purchase_price_list):
    total_purchase_value_list.append(average_purchase_price_list[x] * purchase_count_list[x])
    x +=1
    

# CREATE FINAL DATA FRAME
top_spenders_df = pd.DataFrame({
    "SN": unique_sn_list, "Purchase Count": purchase_count_list, "Average Purchase Price":average_purchase_price_list, "Total Purchase Value":total_purchase_value_list
})

# Sort descending 
top_spenders_df = top_spenders_df.sort_values(by = "Total Purchase Value", ascending = False)

# Finish formatting avg purchase price and total purchase value
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].apply('${:.2f}'.format)
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].apply('${:.2f}'.format)

# Set index to SN and show first 5 rows
top_spenders_df.set_index("SN").head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# ITEM ID AND ITEM NAME
index_list = purchase_data.groupby(["Item ID","Item Name"]).count().index.values

item_id_list = []
item_name_list = []
for item_value in index_list:
    item_id_list.append(item_value[0])
    item_name_list.append(item_value[1])


# PURCHASE COUNT
purchase_count_list = purchase_data.groupby(["Item ID", "Item Name"]).count().loc[:,"Purchase ID"].tolist()

# ITEM PRICE
item_price_list = purchase_data.groupby(["Item ID", "Item Name"]).mean().loc[:,"Price"].tolist()

# TOTAL PURCHASE VALUE
x = 0
total_purchase_value_list = []
while x < len(purchase_count_list):
    total_purchase_value_list.append(item_price_list[x] * purchase_count_list[x])
    x +=1

# CREATE DATA FRAME
most_popular_item_df = pd.DataFrame({
    "Item ID":item_id_list,
    "Item Name":item_name_list,
    "Purchase Count":purchase_count_list,
    "Item Price":item_price_list,
    "Total Purchase Value":total_purchase_value_list
})

# Sort descending by purchase count
most_popular_item_df = most_popular_item_df.sort_values(by = "Purchase Count", ascending = False)

# Finish formatting item price and total purchase value
most_popular_item_df["Item Price"] = most_popular_item_df["Item Price"].apply('${:.2f}'.format)
most_popular_item_df["Total Purchase Value"] = most_popular_item_df["Total Purchase Value"].apply('${:.2f}'.format)


# Set index to SN and show first 5 rows
most_popular_item_df = most_popular_item_df.set_index(["Item ID", "Item Name"])
most_popular_item_df.head()
```
